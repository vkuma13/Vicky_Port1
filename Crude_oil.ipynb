```python
Crude_oil_raw_data['Open']=Crude_oil_raw_data['Open'].abs()
Crude_oil_raw_data['Close']=Crude_oil_raw_data['Close'].abs()
Crude_oil_raw_data['Low']=Crude_oil_raw_data['Low'].abs()
Crude_oil_raw_data['Close']=Crude_oil_raw_data['Close'].abs()

```

```python
Crude_oil_raw_data=Crude_oil_raw_data.drop(['Adj Close'],axis=1)
```

```python


Crude_oil_raw_data['Year']=pd.DatetimeIndex(Crude_oil_raw_data['Date']).year
Crude_oil_raw_data['Month']=pd.DatetimeIndex(Crude_oil_raw_data['Date']).month
Crude_oil_raw_data['Day']=pd.DatetimeIndex(Crude_oil_raw_data['Date']).day
Crude_oil_raw_data
```

```python
Orginal = []
def mon_Naming():
  Crude_oil_raw_data['Month']=pd.DatetimeIndex(Crude_oil_raw_data['Date']).month
  for i in Crude_oil_raw_data['Month']:
    K =calendar.month_name[i]
    Orginal.append(K)
  Crude_oil_raw_data['Month']= Orginal
mon_Naming()
Crude_oil_raw_data
```

```python


#Finding the 107 Null entries 

Crude_oil_raw_data.isnull().sum()

```

```python
# volume of Crude oil 🛢 in liters defining it as a Barrel 
# 1 Barrel of Crude Oil is 159 Liters
sign = []
One_barrel = 159 
for i  in Crude_oil_raw_data['Volume']:
  k = round(i /One_barrel)
  sign.append(k)
Crude_oil_raw_data['Barrels']=sign
Crude_oil_raw_data
```
